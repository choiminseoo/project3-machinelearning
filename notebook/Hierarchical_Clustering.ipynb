```python
import pandas as pd

df=pd.read_csv('행정구역_인구_출근발생량_전기차_충전소_앵커시설_완속.csv')
df.head()
```

### 행정구역 컬럼 제외

```python
col_names=['총 인구수', '총출근발생량', '전기차 수', '앵커시설', '충전소 개수', '완속']
```

```python
x=df[col_names]
x
```

### Standard Scaling 

```python
from sklearn.preprocessing import StandardScaler

# StandardScaler 객체 생성
scaler = StandardScaler()

# 데이터를 StandardScaler에 fitting하고 변환
scaled_data = scaler.fit_transform(x)

# 변환된 데이터 출력
x_scaled = pd.DataFrame(scaled_data, columns=col_names)
x_scaled
```
