```python
import pandas as pd

df=pd.read_csv('/content/drive/MyDrive/XX401/data/전기차충전소_유형.csv')
df.head()
```

```python
df=df[['소재지도로명주소']]
df
```

```python
def sggnm(x):
  a=x.find('구 ')

  if a>=0:
    return x[:a+1]
  else:
    a=x.find('군 ')
    if a>=0:
      return x[:a+1]
    else:
      a=x.find('시 ')
      return x[:a+1]
```

```python
df['시도군']=df['소재지도로명주소'].apply(sggnm)
```

```python
group_df=df.groupby('시도군').count()
group_df
```

```python
indexs=group_df.index
data=group_df.values
indexs=indexs[1:-2]
data=data[1:-2]
print(indexs)
print(data)
```

```python
data = data.reshape(-1)
data
```

```python
df_for_plot={'시도군':indexs,'합계':data}
df_for_plot=pd.DataFrame(df_for_plot)
df_for_plot
```

```python
def address(x):
  if x.find('경기')>=0:
    a=x.find(" ")
    return x[a+1:]
```

```python
df_for_plot['시도군']=df_for_plot['시도군'].apply(address)
df_for_plot
```

```python
df_for_plot['시도군']=df_for_plot['시도군'].str.replace(' ','')
df_for_plot
```

```python
import json
import folium

with open('/content/drive/MyDrive/XX401/data/HangJeongDong_ver20230701.geojson') as response:
    geo_json = json.load(response)

# 경기도 시군구 경계 정보를 가진 geo-json 파일 불러오기
# 경기도 행정구역 경계 지리 정보를 사용

geo_json
```

```python
from sklearn.preprocessing import MinMaxScaler

mMscaler = MinMaxScaler(feature_range=(0,5))
mMscaler.fit(df_for_plot[['합계']])

df_for_plot['합계 scaler'] = mMscaler.transform(df_for_plot[['합계']])
df_for_plot
```

```python
# 경기도 지도 만들기
g_map = folium.Map(location=[37.5502, 126.982], zoom_start=9)
g_map
```

```python
# Choropleth 레이어를 만들고, 맵 m에 추가.

folium.Choropleth(
    geo_data=geo_json,
    data=df_for_plot,
    columns=['시도군', '합계'],
    key_on="properties.sggnm",
    nan_fill_color='white',
    nan_fill_opacity=0.0,
    line_opacity=0.0,
    fill_opacity=0.9
).add_to(g_map)
g_map

#g_map.save("index.html")
```
