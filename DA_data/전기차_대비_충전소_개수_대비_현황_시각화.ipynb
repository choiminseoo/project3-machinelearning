```python
import pandas as pd

df1=pd.read_csv('/content/drive/MyDrive/XX401/data/전기차충전소_유형.csv')
df2=pd.read_csv('/content/drive/MyDrive/XX401/data/연료별용도별자동차등록현황.csv',encoding='cp949')
```

```python
df3=df2[(df2['연료별']=='전기')&(df2['등록연도']==2022)]
df3=df3[['승용차수','승합차수','특수차수','화물차수']].sum().reset_index()
df3[0].sum()
```

```python
print(len(set(df1['충전소명'])))
print(len(df1))
```

```python
data_for_plot=pd.DataFrame()
data_for_plot['합계']=[df3[0].sum(),len(set(df1['충전소명']))]
data_for_plot['index']=['전기차','충전소']
data_for_plot
```
