```python
def cat(x):
  str_list1=['주유소' , '에너지']
  str_list2=['병원','도서관', '센터', '캠핑장', '수련관', '복지', '쉼터', '리조트', '박물관', '관리사무소', '의료원', '문화재단',
             '아람누리','교회', '관리소', '킨텍스', '장례식장', '마리나', '컨벤션', '아트리움', '시니어클럽','센트럴 돔','주민센터','삼융시네마']
  str_list3=['공원', '파크', '왕릉', '유원지', '랜드','공영','시장']
  str_list4=['휴게소', '졸음쉼터', '도로', '터미널']
  str_list5=['회관', '공사', '사무소', '시청', '구청', '기술원', '공단', '환경청', '지사', '테크노파크', '군청', '교육지원청', '수덕원', '교육원', '하수처리장', '연수원', '연구원'
             '진흥원', '중부지소', '경찰서', '소방서', '경챁청', '예비군', '사업단', '교도소', '파출소' , '지구대', '보건소', '보건지소']
  str_list6=['이마트', '스타필드', '마트', '아울렛', '홈플러스', '농협', '벨라시티', '가게', '편의점', '매장']
  str_list7=['학교', '주차', '주차장', '대학', '충전소']
  str_list8=['영업소', '사업소', '(주)', '에이팩','스테이션', '엠파이어', '비전타워', 'IT밸리', '운수', '공장',
             '매매', '서비스']
  str_list9=['체육관','체육광장','운동장','스타디움','종합경기타운','레포츠타운','수영장','수영관']


  if any(str in x for str in str_list1) :
    return '주유소'
  elif any(str in x for str in str_list2) :
    return '편의시설'
  elif any(str in x for str in str_list3) :
    return '공영주차장'
  elif any(str in x for str in str_list4) :
    return '도로, 휴게소'
  elif any(str in x for str in str_list5) :
    return '공공기관'
  elif any(str in x for str in str_list6) :
    return '쇼핑센터'
  elif any(str in x for str in str_list7) :
    return '주차장'
  elif any(str in x for str in str_list8) :
    return '상업소'
  elif any(str in x for str in str_list9) :
    return '스포츠, 경기장'
  else :
    return '주거공간'

#충전기타입 - 1: DC차데모, 2: AC완속, 3: DC차데모+AC3상, 4: DC콤보, 5: DC차데모+DC콤보, 6: DC차데모+AC3상+DC콤보, 7: AC3상, 8: DC콤보(완속)
def charger(x):
  n=int(x)
  if n==2 or n==8:
    return '완속'
  elif n==3 or n==5 or n==6:
    return '급속/완속'
  else:
    return '급속'
```

```python
df['분류유형']='주거공간'
df['충전기 타입']='완속'
df['분류유형']= df['충전소명'].apply(cat)
df['충전기 타입']= df['충전기타입'].apply(charger)
df.head()
```

```python
df['주소명'] = df.apply(lambda x: x['소재지도로명주소'] if pd.isna(x['구주소']) else x['구주소'], axis=1)
df1=df[['충전소명','위도','경도','주소명','충전소ID','충전기ID','분류유형','충전기 타입']]
df1
```

```python
def addre(x):
  a=x.split()
  ss=''
  for ax in a:
    if '경기' in ax:
      pass
    elif ax[-1]=='구':
      ss+=ax
    elif ax[-1]=='읍' or ax[-1]=='동' or ax[-1]=='면':
      break
    else:
      ss=ss+' '+ax
  return addre2(ss.strip())

def addre2(x):
  a=x.split()
  ss=''
  for ax in a:
    if ax[-1]=='시' or ax[-1]=='구' or ax[-1]=='군':
      return ax
```

```python
df1['시군구']=df1['주소명'].apply(addre)
df1
```

```python
import pandas as pd
df1=pd.read_csv('/content/drive/MyDrive/XX401/data/전기차충전소_유형_주소_충전기타입.csv')
```

```python
df2=df1
df2=df2[['시군구','충전소명','충전기 타입']].groupby(['시군구','충전기 타입']).count().reset_index()
df2.iloc[2]
```

```python
df2.iloc[2]['충전기 타입']
```

```python
df_merge=pd.DataFrame()
df_merge['시군구']=df1['시군구']
df_merge = df_merge.dropna(axis = 0)
df_merge=df_merge.drop_duplicates(subset='시군구', keep='first',ignore_index=True)
df_merge
```

```python
for index in range(len(df2)):

  if df2.iloc[index]['충전기 타입']=='완속':
    df_merge.loc[df_merge['시군구']==df2.iloc[index]['시군구'],'완속']=df2.iloc[index]['충전소명']
  elif df2.iloc[index]['충전기 타입']=='급속/완속':
    df_merge.loc[df_merge['시군구']==df2.iloc[index]['시군구'],'급속/완속']=df2.iloc[index]['충전소명']
  else:
    df_merge.loc[df_merge['시군구']==df2.iloc[index]['시군구'],'급속']=df2.iloc[index]['충전소명']
```

```python
df_merge['급속']=df_merge['급속']+df_merge['급속/완속']
df_merge=df_merge[['시군구','급속','완속']]
df_merge
```

```python
df3=df1
df3=df3[['시군구','충전소명','분류유형']].groupby(['시군구','분류유형']).count().reset_index()
df3
```

```python
df_merge=df_merge.fillna(0)
```
