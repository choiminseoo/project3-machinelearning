```python
df=df[df['행정구역구분명']=='읍면동']
df
```

```python
def addre(x):
  a=x.split()
  ss=''
  for ax in a:
    if ax[-1]=='구':
      ss+=ax
    else:
      ss=ss+' '+ax
  return ss.strip()
```

```python
df['행정구역명_2']=df['행정구역명'].apply(addre)
df
```

```python
s='경기도 성남시 분당구'
a=addre(s)
print(a)
```

```python
# 총인구수 minmax scaling
from sklearn.preprocessing import MinMaxScaler

mMscaler = MinMaxScaler(feature_range=(0,5))
mMscaler.fit(df_for_plot[['총 인구수']])

df_for_plot['총 인구수 scaler'] = mMscaler.transform(df[['총 인구수']])
```

```python
import json
import folium

with open('/content/drive/MyDrive/XX401/data/HangJeongDong_ver20230701.geojson') as response:
    geo_json = json.load(response)

# 경기도 시군구 경계 정보를 가진 geo-json 파일 불러오기
# 경기도 행정구역 경계 지리 정보를 사용

geo_json
```

```python
# 경기도 지도 만들기
g_map = folium.Map(location=[37.5502, 126.982], zoom_start=9)
g_map
```

```python
# Choropleth 레이어를 만들고, 맵 m에 추가.

folium.Choropleth(
    geo_data=geo_json,
    data=df_for_plot,
    columns=['행정구역명_2','총 인구수'],
    key_on="feature.properties.adm_nm",
    nan_fill_color='white',
    nan_fill_opacity=0.0,
    line_opacity=0.0,
    fill_opacity=0.9
).add_to(g_map)
g_map

#g_map.save("index.html")
```

```python
import plotly.express as px
fig = px.choropleth_mapbox(data_frame=df_for_plot, geojson=geo_json,
                           featureidkey="properties.adm_nm",
                           locations=df_for_plot['행정구역명_2'], color="총 인구수",
                           opacity=0.5, zoom=9,
                           center={"lat": 37.541, "lon": 126.986},
                           mapbox_style="carto-positron",
                           color_continuous_scale=px.colors.sequential.Cividis_r
                           )
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
#fig.write_html("index.html")
```
