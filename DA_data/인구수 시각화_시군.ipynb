```python
import pandas as pd

df=pd.read_csv('/content/drive/MyDrive/XX401/data/주민등록인구집계현황.csv',encoding='cp949')
df.head()
```

```python
df=df[(df['연도']==2024)&(df['월']==5)]
df
```

```python
df=df[['행정구역구분명','행정구역명','총 인구수']]
print(len(df))
print(df.head())
```

```python
df=df[df['행정구역구분명']=='읍면동']
```

```python
def sggnm(x):
  a=x.find('구 ')

  if a>=0:
    return x[:a+1]
  else:
    a=x.find('군 ')
    if a>=0:
      return x[:a+1]
    else:
      a=x.find('시 ')
      return x[:a+1]

def address(x):
  if x.find('경기')>=0:
    a=x.find(" ")
    return x[a+1:]
```

```python
df2=df
df2['행정구역명']=df['행정구역명'].apply(sggnm)
df2
```

```python
group_df=df2.groupby('행정구역명')['총 인구수'].sum()
group_df
```

```python
indexs=group_df.index
data=group_df.values
data = data.reshape(-1)
print(indexs)
print(data)
```

```python
df_for_plot={'시도군':indexs[1:],'합계':data[1:]}
df_for_plot=pd.DataFrame(df_for_plot)
df_for_plot
```

```python
df_for_plot['시도군']=df_for_plot['시도군'].apply(address)
df_for_plot
```

```python
df_for_plot['시도군']=df_for_plot['시도군'].str.replace(' ','')
df_for_plot
```

```python
import json
import folium

with open('/content/drive/MyDrive/XX401/data/HangJeongDong_ver20230701.geojson') as response:
    geo_json = json.load(response)

# 경기도 시군구 경계 정보를 가진 geo-json 파일 불러오기
# 경기도 행정구역 경계 지리 정보를 사용

geo_json
```

```python
# 경기도 지도 만들기
g_map = folium.Map(location=[37.5502, 126.982], zoom_start=9)
g_map
```

```python
# 총인구수 minmax scaling
from sklearn.preprocessing import MinMaxScaler

mMscaler = MinMaxScaler(feature_range=(0,5))
mMscaler.fit(df_for_plot[['합계']])

df_for_plot['합계 scaler'] = mMscaler.transform(df_for_plot[['합계']])
df_for_plot
```

```python
import plotly.express as px

folium.Choropleth(
    geo_data=geo_json,
    data=df_for_plot,
    columns=['시도군', '합계 scaler'],
    key_on="properties.sggnm",
    nan_fill_color='white',
    nan_fill_opacity=0.0,
    line_opacity=0.0,
    fill_opacity=1.0
).add_to(g_map)
g_map
```
