```python
import pandas as pd

df=pd.read_csv('/content/drive/MyDrive/XX401/data/연료별용도별자동차등록현황.csv',encoding='cp949')
df.head()
```

```python
df=df[(df['등록연도']==2022)&(df['연료별']=='전기')]
df
```

```python
group_df=df.groupby('시군구명')[['승용차수','승합차수','특수차수','화물차수']].sum().reset_index()
group_df
```

```python
group_df['전기차 합계']=group_df['승용차수']+group_df['승합차수']+group_df['특수차수']+group_df['화물차수']
df_for_plot=group_df[['시군구명','전기차 합계']]
df_for_plot
```

```python
import json
import folium

with open('/content/drive/MyDrive/XX401/data/HangJeongDong_ver20230701.geojson') as response:
    geo_json = json.load(response)

# 경기도 시군구 경계 정보를 가진 geo-json 파일 불러오기
# 경기도 행정구역 경계 지리 정보를 사용

geo_json
```

```python
df_for_plot['시군구명']=group_df['시군구명'].str.replace(' ','')
df_for_plot
```

```python
df_for_plot['전기차 합계']=group_df['승용차수']+group_df['승합차수']+group_df['특수차수']+group_df['화물차수']
df_for_plot
```

```python
# 전기차 수 minmax scaling
from sklearn.preprocessing import MinMaxScaler

mMscaler = MinMaxScaler(feature_range=(0,5))
mMscaler.fit(df_for_plot[['전기차 합계']])

df_for_plot['전기차 합계 scaler'] = mMscaler.transform(df_for_plot[['전기차 합계']])
df_for_plot
```

```python
import plotly.express as px
fig = px.choropleth_mapbox(data_frame=df_for_plot, geojson=geo_json,
                           featureidkey="properties.sggnm",
                           locations=df_for_plot['시군구명'], color="전기차 합계 scaler",
                           opacity=0.5, zoom=9,
                           center={"lat": 37.541, "lon": 126.986},
                           mapbox_style="carto-positron"
                           )
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
#fig.write_html("index.html")
```

```python
# 경기도 지도 만들기
g_map = folium.Map(location=[37.5502, 126.982], zoom_start=9)
g_map
```

```python
# Choropleth 레이어를 만들고, 맵 m에 추가.

folium.Choropleth(
    geo_data=geo_json,
    data=df_for_plot,
    columns=['시군구명', '전기차 합계 scaler'],
    key_on="properties.sggnm",
    nan_fill_color='white',
    nan_fill_opacity=0.0,
    line_opacity=0.0,
).add_to(g_map)
g_map

#g_map.save("index.html")
```
