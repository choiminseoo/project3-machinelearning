```python
import pandas as pd
df=pd.read_csv('C:/Users/yellni/Desktop/XX401/Data/전기차충전소_유형.csv')

df.head()
```

```python
def cat(x):
    str_list1=['체육관','체육광장','운동장','스타디움','종합경기타운','레포츠타운','수영장','수영관']
    str_list2=['이마트', '스타필드', '마트', '아울렛', '홈플러스', '농협', '벨라시티', '가게', '편의점', '매장']
    str_list3=['공원', '파크', '왕릉', '유원지', '랜드']
    str_list4=['삼융시네마']


    if any(str in x for str in str_list1) :
        return '스포츠경기장'
    elif any(str in x for str in str_list2) :
        return '대형쇼핑몰'
    elif any(str in x for str in str_list3) :
        return '공원'
    elif any(str in x for str in str_list4) :
        return '영화관'
    else :
        return '편의시설 외'
```

```python
df['분류유형']='편의시설'


df['분류유형']= df['충전소명'].apply(cat)
df.head()
```

```python
df.to_csv('C:/Users/yellni/Desktop/XX401/Data/전기차충전소_편의시설.csv',index=False)
```

```python
def get_color(x):
    if x=='공원':
        return 'orange'
    elif x=='대형쇼핑몰':
        return 'blue'
    elif x=='스포츠경기장':
        return 'yellow'
    elif x=='영화관':
        return 'green'
    #elif x=='편의시설 외':
        #return 'pink'
```

```python
import folium

m = folium.Map(
    location=[37.559819, 126.963895],
    zoom_start=11,
)

df['위도'].apply(lambda x: float(x))
df['경도'].apply(lambda x: float(x))


for index in range(len(df)):
    color=get_color(df.iloc[index]['분류유형'])
    folium.Circle(radius=2,location=[df.iloc[index]['위도'], df.iloc[index]['경도']], color=color).add_to(m)

#mymap.save("map_with_colored_markers.html")
m
```
